Now implement that into Spider logic, with pagnation, which loads items while there is button to load more present.

Output will be saved in nivea_products.json

```python
import nest_asyncio
nest_asyncio.apply()

import scrapy
from urllib.parse import urljoin
from scrapy.crawler import CrawlerProcess


class NiveaSpider(scrapy.Spider):
    name = 'nivea'
    start_urls = ['https://www.nivea.hr/proizvodi']
    
    def parse(self, response):
        for prod in response.css('article.nx-product-teaser--compact'):
            item = {
                'name':         prod.attrib.get('data-name', '').strip(),
                'category':     prod.attrib.get('data-category', '').strip(),
                'url':          response.urljoin(prod.css('a::attr(href)').get()),
                'rating':       prod.css('span.nx-rating__total::attr(data-rating)').get(default='0'),
                'rating_count': prod.css('span.nx-rating__count::text').get(default='(0)').strip('()'),
            }
            print("Getting information for: ", item["url"])
            yield item

        next_path = response.css('button.nx-btn--load-more::attr(data-ajax-url)').get()
        if next_path:
            next_url = response.urljoin(next_path)
            yield scrapy.Request(next_url, callback=self.parse)

process = CrawlerProcess(settings={
    'LOG_LEVEL': 'WARNING',
    'FEEDS': {
        'nivea_products.json': {
            'format': 'json',
            'encoding': 'utf8',
            'indent': 2,
        },
    },
})

process.crawl(NiveaSpider)
process.start()
```
